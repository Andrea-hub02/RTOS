### <span style="color:blue">Problem statement</span>

Starting with the operating systems or so with their acronyms (OS) in servers, it is taken into account that the servers serve to give answer to one or more clients. 
the servers have a great advantage since this can help us that the information can be backed up and even also that is not lost time in the company, or to take care of that information is not lost although the hardware is damaged, well thus for that reason the servers are important.

The operating systems (OS), can help to be compatible with the servers, not all the OS are compatible, since MacOs, Windows, does not have compatiblidad and Unix if it has it.

In this project it will be implemented in Kubernetes servers because it can allow us to accredit what we need in this project.
Kubernetes is a modern application that allows us to use containers, it allows us to create packaged microservices with their dependencies and configurations, since Kubernetes is an open source software that implements and manages containers.

The project uses FreeRTOS CMSIS Ver. 2. The scheduling policy is Rate Monotonic (RM) with time triggering.
